```python
from sklearn.cluster import KMeans
```

```python
X=df[["Retweets","Likes"]]
X
```

```python

X.plot.scatter(x='Retweets', y='Likes')
```

```python
km=KMeans(1)
km.fit(X)
X["cl"]=km.labels_
X.plot.scatter(x='Retweets', y='Likes',c="cl",colormap='gist_rainbow')
```
